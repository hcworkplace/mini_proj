토큰화 + 불용어제거

```
encoder_input, decoder_input, decoder_output = [], [], []
```

```
mecab = Mecab()

# 불용어 정의
stopwords=['인데요','다','을','습니다','Q','이다','보인다','한다','했다','였다','이다','의','가','이','은','들','는','좀','잘','과','도','를','으로','자','에','와','하다']

for line in data['content']:
    temp = []
    temp = mecab.morphs(str(line))
    temp = [word for word in temp if not word in stopwords] # 불용어 제거
    encoder_input.append(temp)
```

```
for line in data['title']:
    temp = []
    temp.append('<start>')
    for word in mecab.morphs(str(line)):
      if word not in stopwords: # 불용어 제거
        temp.append(word) 
    decoder_input.append(temp)
```

```
for line in data['title']:
    temp = []
    for word in mecab.morphs(str(line)):
      if word not in stopwords: # 불용어 제거
        temp.append(word) 
    temp.append('<end>')
    decoder_output.append(temp)
```

정수인코딩

```
import nltk
from nltk import FreqDist
import numpy as np

# 단어-빈도수 조합으로 이루어진 단어집합 해시맵 vocab을 생성한다.
# NLTK에서는 빈도수 계산 도구인 FreqDist()를 지원한다.
vocab = FreqDist(np.hstack(encoder_input))
print('단어 [코로나]의 빈도수는? ', vocab['코로나'], '번')

# most_common(N) : 가장 빈도수가 높은 N개의 단어를 반환
# 상위 1000000개의 단어만 보존
vocab = vocab.most_common(10000000)
vocab
```

```
word_to_index = {}

# 단어들에 순차적으로 1 ~ 까지의 인덱스를 부여한다
word_to_index = {word[0] : index + 1 for index, word in enumerate(vocab)}

# 특수 인덱스
word_to_index['unk'] = 0

encoded_input = []
for line in encoder_input:
  tmp = []
  for word in line:
    try:
      # 각 글자를 해당하는 정수로 변환한다.
      tmp.append(word_to_index[word])
    except KeyError: 
      # 단어 집합에 없는 단어일 경우(=빈도수 상위 100000 이외의 단어) unk로 대체된다.
      tmp.append(word_to_index['unk'])

  encoded_input.append(tmp)

# 기존의 리뷰가 성공적으로 encoding 되었는지 확인해보기
print(encoded_input[1])
print(len(encoded_input[1]))
print(len(data['content'][1]))
```

```
decoded_input = []
for line in decoder_input:
  tmp = []
  for word in line:
    try:
      # 각 글자를 해당하는 정수로 변환한다.
      tmp.append(word_to_index[word])
    except KeyError: 
      # 단어 집합에 없는 단어일 경우(=빈도수 상위 100000 이외의 단어) unk로 대체된다.
      tmp.append(word_to_index['unk'])

  decoded_input.append(tmp)
```

```
decoded_output = []
for line in decoder_output:
  tmp = []
  for word in line:
    try:
      # 각 글자를 해당하는 정수로 변환한다.
      tmp.append(word_to_index[word])
    except KeyError: 
      # 단어 집합에 없는 단어일 경우(=빈도수 상위 100000 이외의 단어) unk로 대체된다.
      tmp.append(word_to_index['unk'])

  decoded_output.append(tmp)

# 기존의 리뷰가 성공적으로 encoding 되었는지 확인해보기
print(encoded_input[0])
print(decoded_input[0])
print(decoded_output[0])
# print(len(encoded_input[0]))
# print(len(data['content'][0]))
```
